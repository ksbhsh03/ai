```python
import tensorflow as tf
import numpy as np

import pandas as pd
import random
import os
import numpy as np
import math

from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
class CFG:
    SEED = 42
```

```python
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
seed_everything(CFG.SEED) # Seed 고정
```

```python
train = pd.read_csv('open/train.csv')
test = pd.read_csv('open/test.csv')
```

```python
def get_x_y(df):
    if 'class' in df.columns:
        df_x = df.drop(columns=['id', 'class'])
        df_y = df['class']
        return df_x, df_y
    else:
        df_x = df.drop(columns=['id'])
        return df_x
```

```python
train_x, train_y = get_x_y(train)
test_x = get_x_y(test)
```

```python
class_le = preprocessing.LabelEncoder()
snp_le = preprocessing.LabelEncoder()
snp_col = [f'SNP_{str(x).zfill(2)}' for x in range(1,16)]
```

```python
snp_data = []
for col in snp_col:
    snp_data += list(train_x[col].values)
```

```python
train_y = class_le.fit_transform(train_y)
snp_le.fit(snp_data)
```

```python
for col in train_x.columns:
    if col in snp_col:
        train_x[col] = snp_le.transform(train_x[col])
        test_x[col] = snp_le.transform(test_x[col])
```

```python
np_classes = 15

# utilts안에 one_hot을 지원해주고있다
y_one_hot = tf.keras.utils.to_categorical(train_y, np_classes)
y_one_hot
```

```python
train_x.astype("uint8")
print(train_x)
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=train_x.shape[1], units=y_one_hot.shape[1], activation='softmax'))
```

```python
tf.model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
                                                                                    metrics=['accuracy'])
tf.model.summary()
```

```python
tf.model.fit(train_x, y_one_hot, epochs=1000) # , verbose=0 
```

```python
test_data = test_x
preds = tf.model.predict(test_data)
preds.argmax(axis=1)
preds.shape
```

```python
submit = pd.read_csv('open/sample_submission.csv')
```

```python
submit['class'] = class_le.inverse_transform(preds)
```

```python
submit.to_csv('./answer/categorical.csv', index=False)
```
