```python
import pandas as pd
import random
import os
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from xgboost import XGBClassifier
```

```python
train = pd.read_csv('open/train.csv')
test = pd.read_csv('open/test.csv')
```

```python
def get_x_y(df):
    if 'class' in df.columns:
        df_x = df.drop(columns=['id', 'class'])
        df_y = df['class']
        return df_x, df_y
    else:
        df_x = df.drop(columns=['id'])
        return df_x
```

```python
train_x, train_y = get_x_y(train)
test_x = get_x_y(test)
```

```python
# 0인 데이터 제거
train.drop(['mother','father','id','gender'], axis=1,inplace=True)
```

```python
test.drop(['mother','father','id','gender'], axis=1,inplace=True)
```

```python
class_le = preprocessing.LabelEncoder()
snp_le = preprocessing.LabelEncoder()
snp_col = [f'SNP_{str(x).zfill(2)}' for x in range(1,16)]
```

```python
snp_data = []
for col in snp_col:
    snp_data += list(train_x[col].values)
```

```python
train_x.headtrain_y = class_le.fit_transform(train_y)
snp_le.fit(snp_data)
```

```python
def cross_validation(classifier, train_x, train_y):
    cv_scores =[]
    
    for i in range(10):
        scores = cross_val_score(classifier, train_x, train_y, cv=50, scoring='accuracy')  # 훈려데이터가 적을때 많이 사용된다. 
        cv_scores.append(scores.mean())
        
    return cv_scores
```

```python
rt_cv_scores = cross_validation(XGBClassifier(), train_x, train_y)
```

```python
for col in train_x.columns:
    if col in snp_col:
        train_x[col] = snp_le.transform(train_x[col])
        test_x[col] = snp_le.transform(test_x[col])
```

```python
clf = XGBClassifier(n_estimators=400, learning_rate=0.02, max_depth=5)
clf.fit(train_x, train_y)
```

```python
preds = clf.predict(test_x)
print(preds)
```
