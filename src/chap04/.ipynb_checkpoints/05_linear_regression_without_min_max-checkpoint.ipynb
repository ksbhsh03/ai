```python
import tensorflow as tf
import numpy as np

# open시작값, hight최고가, volume거래량, low최저가, close종료값
xy = np.array([[828.659973, 833.450012, 908100, 828.349976, 831.659973],
               [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
               [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
               [816, 820.958984, 1008100, 815.48999, 819.23999],
               [819.359985, 823, 1188100, 818.469971, 818.97998],
               [819, 823, 1198100, 816, 820.450012],
               [811.700012, 815.25, 1098100, 809.780029, 813.669983],
               [809.51001, 816.659973, 1398100, 804.539978, 809.559998]]) 
```

```python
# 회귀모델
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=x_data.shape[1], units=1))
tf.model.add(tf.keras.layers.Activation('linear'))  # 회귀모델을 만들떄는 Activation에서의 매개변수는 linear
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-10))
tf.model.summary()
```

```python
loss = tf.model.fit(x_data, y_data, epochs=100)
```
