```python
import matplotlib.pyplot as plt  # python에서는 plt라고 별칭을 사용한다
import numpy as np
```

```python
# 단순한 그래프 그리기

# 데이터 준비
x = np.arange(0, 6, 0.1)  # arange는 반환된 값이 matrix값이다.
y1 = np.sin(x)            # sin
y2 = np.cos(x)            #
```

```python
# 그래프 그리기

plt.plot(x, y1, label = 'sin')              # lable매개변수는 고유이름을 가져간다.
plt.plot(x, y2, linestyle = '--', label = 'cos')           #  linestyle = '--'  --두개는 라인을 점선으로 표현하는 의미이다.
plt.xlabel('x')          # x축의 이름
plt.ylabel('y')          # y축의 이름
plt.title('sin & cos')   #제목
plt.legend()             # 범례   label이름을가져와서 출력
plt.show()
```

```python
#np.random.seed(1234)  #seed()는 난수를 값을 발생시켜서 고정.
x = np.arange(10)
y = np.random.rand(10) # 0 ~ 1 사이의 균일한 분포

plt.plot(x, y)  # 꺽은선 그래프를 등록
plt.show()  # show() 시각화보여주는 함수
```

```python
# 3차 함수 f(x) = (x-2) * x * (x+2)

def f(x):
    return (x-2) * x * (x+2)
print(f(0))
print(f(2))
print(f(-2))
```

```python
# x 값에 대해 ndarray 배열이며 각각에 대응한 f를 한꺼번에 ndarray로 돌려줌.
print(f(np.array([1, 2, 3])))  #배열로 값을 전달하면 몇개를 전달하든 값을 배열로다시 return해서 전달해준다.
```

```python
# 그래프를 그리는 x의 범위를 -3 ~ +3 까지로 하고. 간격 0.5
x = np.arange(-3, 3.5, 0.5)
plt.plot(x, f(x))
plt.show()
```

```python
# 그래프를 장식

def f2(x, w):
    return (x-w) * x * (x + 2)   # 함수 정의
```

```python
# x를 정의
x = np.linspace(-3, 3 , 100)  # x를 100분할하기
```

```python
# 차트묘사
plt.plot(x, f2(x, 2), color='black', label='x')   
plt.plot(x, f2(x, 1), color='blue', label='y')
plt.legend(loc = 'upper left')    # legend()사용할때는 위에서 label매개변수를 지정해줘야 실행된다.
plt.ylim(-15, 15)  # y축의 범위 지정
plt.title('f2(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)     # 그리드(눈금)
plt.show
```

```python
# 그래프를 여러 개 보여주기

plt.figure(figsize=(10, 3)) # 값을 전달해서 전체 영역의 크기를 지정  # 2번째 매개변수가 기준이된다
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)  # wspace넓이간격 hspace 높이간격
for i in range(6):
    plt.subplot(2, 3, i+1)    # 2행 3열은 1부터 그래프 위치를 지정
    plt.title(i+1)
    plt.plot(x, f2(x, i))
    plt.ylim(-20, 20)
    plt.grid(True)
    
plt.show()  # show()는 한번만 실행하면 된다 (반복문 포함)
```
