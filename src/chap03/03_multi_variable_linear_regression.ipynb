```python
import tensorflow as tf
import numpy as np
```

```python
x_data = [[73., 80., 85.],   # 배열로 바꼈을때  2차원 5행 3열 
          [93., 88., 93.],
          [89., 91., 90.],  
          [96., 98., 100.],
          [73., 66., 70.]
         ]
y_data = [[152.],   # 배열로 바꼈을때 2차원 5행 1열
          [185.],
          [180.],
          [196.],
          [142.]]
```

```python
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(input_dim=3, units=1))
tf.model.add(tf.keras.layers.Activation('linear'))  # 활성화 함스
```

```python
#_________________________________________________________________
#Layer (type)                Output Shape              Param #   
#=================================================================
# dense (Dense)               (None, 1)                 4  찾아야할 W값 + b
#
                                                                 


tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5))
tf.model.summary()
```

```python
loss = tf.model.fit(x_data, y_data, epochs=200)
```

```python
y_predict = tf.model.predict(np.array([[72.0, 93.0, 90.]]))
y_predict
```
