```python
import tensorflow as tf
import numpy as np
```

```python
#loadtxt()는 파일 읽어오게하는 함수 
# delimiter 구분자
# dtype 타입을 알려줘야한다.
xy = np.loadtxt('../data/data-01-test-score.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1] 
y_data = xy[:, [-1]] 
```

```python
print(x_data,'\nx_data shape:', x_data.shape)
print(y_data,'\ny_data shape:', y_data.shape)
```

```python
tf.model = tf.keras.Sequential()

# activation 활성화함수는 제일 마지막에만 붙여한다.
# 제일앞에 대문자면 참조자료형
# 매개변수는 스스로 new해주고있다는 의미이다.
tf.model.add(tf.keras.layers.Dense(input_dim=3, units=1, activation='linear'))
```

```python
tf.model.summary()
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5))
```

```python
tf.model.fit(x_data, y_data, epochs=3500)
```

```python
print(tf.model.predict([[100., 70., 98.]]))
```
